```python
df_addr = pd.read_excel("./KPMG_VI_New_raw_data_update_final.xlsx", sheet_name="CustomerAddress", header=1)
df_addr.head()
```

```python
df_demo = pd.read_excel("./KPMG_VI_New_raw_data_update_final.xlsx", sheet_name="CustomerDemographic", header=1)
df_demo.head()
```

```python
# combine customer address and customer demographic, with outer join.
demo_addr = pd.merge(df_demo, df_addr, on='customer_id', how='outer')
print(f"shape:{demo_addr.shape}")
```

### using indicator param in merge, in customer demo and customer address tables.

```python
# using indicator.
indicator = pd.merge(df_addr, df_trans, on='customer_id', how='outer', indicator=True)
indicator.head()
```

```python
_merge = indicator.loc[ indicator['_merge'] != 'both' ]
_merge.head()
```

```python
# use indicator. check for customer_id in transactions and demographic table.s
demo_trans = pd.merge(df_demo, df_trans, on='customer_id', how='outer', indicator=True)
# demo_trans.loc[ demo_trans['_merge'] != "both" ]
print(f"{demo_trans.shape}")
demo_trans.head()
```
